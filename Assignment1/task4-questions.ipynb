   Your answer: **[
   
Use BGD, SGD to represent Stochastic Gradient Descent and Batch Gradient Descent respectively.

1.BGD shuffles the data sample, while a random order is not required in SGD.

2.BGD is not suitable for large training samples while SGD is usually used for large training samples.

3.BGD computes gradient using the whole Training sample, while SGD computes gradient using a single training sample.

4.BGD usually descends slower and computationally more expensive than SGD.

5.BGD is able to gives optimal solution with sufficient time to converge while SGD tends to gives good solution but not optimal.

6.BGD can't easily escape from local minimum while SGD gets a bigger chance to escape .

Since BGD and SGD both have advantages and drawbacks, implement a mixture of BGD and SGD is one of the best solutions in real-world problems and that is called Mini-batch GD. It uses a batch of a fixed number of training examples which is less than the actual dataset, called mini-batch, by doing so, it is equipped with the advantages of both the BGD and SGD, allowing us to update our parameters frequently and use vectorized implementation for more efficient computations.

   Your answer: **[
   
Why are activation functions used in deep learning?
Not all the information is equally useful by activation functions we are able to get rid of certain useless features so that the whole neural network will be more efficient. A neural network without an activation function is essentially just a linear regression model. Although linear transformations make the neural network simpler, but this network would be less powerful and will not be able to learn the complex patterns from the data. 

For values greater than 3 or less than -3, Sigmoid will have very small gradients. As the gradient value approaches zero, the network is not really learning. Additionally, the sigmoid function is not symmetric around zero. So output of all the neurons will be of the same sign, though this could be fixed by scaling the sigmoid function.

The gradient of the tanh function is steeper as compared to the sigmoid function, while neural networks can still be influenced as its gradient values approaches zero, as the result, neural networks will almost stop learning.
